# Forcasts for total CO2 emissions in Europe

In this notebook we will be visualizing the forecasts for total CO2 emissions in Europe. 

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Data loading

```python
data_root = './data/'
```

```python
# Reading the data
df_co2 = pd.read_csv(f'{data_root}co2_emissions_transformed.csv', sep=';')
df_co2_sarima = pd.read_csv(f'{data_root}forecasts_sarima.csv', sep=';')
```

```python
df_past_grouped = df_co2.groupby('Year')['CO2_emissions'].sum().reset_index()
df_pred_grouped = df_co2_sarima.groupby('year')['CO2_emissions'].sum().reset_index()
```

```python
plt.figure(figsize=(16, 6))
plt.grid(axis="y", linestyle="--", alpha=0.5, zorder=0)
plt.bar(df_past_grouped['Year'], df_past_grouped['CO2_emissions'], label='Actual', color='dodgerblue')
plt.bar(df_pred_grouped['year'], df_pred_grouped['CO2_emissions'], label='Prediction', color='orange')
plt.title('Forcasts for total CO2 emissions in Europe')
plt.xlabel("Year")
plt.ylabel("CO Emissions")
plt.legend()
plt.tight_layout()
plt.show()
```

Here is the grouped overall CO2 emissions for Europe's countries that we had in the dataset. We can see a decline of CO2 emissions in the near future.
